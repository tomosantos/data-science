```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing

%matplotlib inline
```

```python
dataset = fetch_california_housing()
dataset
```

```python
dataset.keys()
```

```python
print(dataset.DESCR)
```

```python
dataset.target_names
```

```python
## Preparing the DataFrame
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
df['Price'] = dataset.target
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.corr()
```

```python
## Independent and Dependent Features
X = df.iloc[:, :-1]     # Independent Features
y = df.iloc[:, -1]      # Dependent Features

X.head()
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=10)
```

```python
X_train.shape, X_test.shape
```

```python
## Standardize Data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

## Model Training

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model
```

```python
model.fit(X_train, y_train)
```

```python
# Slope or Coefficients
model.coef_
```

```python
# Intercepts
model.intercept_
```

```python
# Prediction for the test data
y_pred = model.predict(X_test)
y_pred
```

```python
# Performance Metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# R squared
r2 = r2_score(y_test, y_pred)
print(r2)
```

```python
#display adjusted R-squared
1 - (1-r2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

## Assumptions

```python
plt.scatter(y_test, y_pred)
plt.xlabel('Test Truth Data')
plt.ylabel('Test Predicted Data')
```

## Pickling

Python pickle module is used for serialising and de-serialising a Python object structure. Any object in Python can be pickled so that it can be saved on disk. What pickle does is that it 'serialises' the object first before writint it to file. Pickling is a way to convert a Python object (list, dict, etc.) into a character stream. The idea is that this character stream contains all the information necessary to reconstruct the object in another Python script

```python
import pickle

pickle.dump(model, open('data/lin_regressor.pkl', 'wb'))
```

```python
model = pickle.load(open('data/lin_regressor.pkl', 'rb'))
```
