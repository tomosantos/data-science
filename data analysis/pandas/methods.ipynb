## Grouby

```python
import pandas as pd

data = {
    'Classe': ['Júnior', 'Júnior', 'Pleno', 'Pleno', 'Sênior', 'Sênior'],
    'Nome': ['Jorge', 'Carlos', 'Roberta', 'Patrícia', 'Bruno', 'Vera'],
    'Venda': [200, 120, 340, 124, 243, 350]
}

df = pd.DataFrame(data)
```

```python
df.groupby('Classe').sum()
```

```python
df.groupby('Classe').min()
```

```python
df.groupby('Classe').max()
```

```python
df2 = df.copy()
```

```python
df2['Venda'] = [150, 432, 190, 230, 410, 155]
```

```python
df
```

```python
df2
```

```python
df3 = pd.concat([df, df2])
df3.groupby(['Classe', 'Nome']).sum()
```

## Merge, Join and Concat

### Concat

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
pd.concat([df1, df2, df3], axis='rows') # default
```

```python
pd.concat([df1, df2, df3], axis='columns')
```

Se o `DataFrame` compartilha índices, então usaremos `axis=columns`. Agora, se compartilha colunas, então usaremos `axis=rows`.

### Merge

```python
esquerda = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
direita = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda, direita)
```

No entanto, a maneira mais correta seria especificar o ponto de *merge*.

```python
pd.merge(esquerda, direita, on='key')
```

```python
esquerda = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
direita = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda, direita, how='inner', on=['key1', 'key2']) # as colunas k1 e k2 da esquerda tentarão buscar por associações correspondentes nas colunas k1 e k2 da direita.
                            # default
```

```python
pd.merge(esquerda, direita, how="outer", on=['key1', 'key2']) # caso não haja associações correspondentes nos dois df's, pandas preencherá os campos com 'NaN'.
```

### Join

```python
esquerda = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

direita = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
esquerda
```

```python
direita
```

```python
esquerda.join(direita)
```
