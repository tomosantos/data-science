3. Seja a matriz de dados X derivada do dataframe:

```python
# subselecionar variáveis e transformar em matriz
X = df.values[:, 1:]
X = np.asarray(X, dtype=float)
X
```

```python
X.shape
```

```python
De = squareform(pdist(X, metric='euclidean'))
Dep = squareform(pdist(X, metric='seuclidean'))
Dm = squareform(pdist(X, metric='mahalanobis'))
```

```python
def compara_dist(df, op='max'):
    if op == 'max':
        maximo = np.amax(df.values)
        a = np.where(df.values == maximo)
        valor = maximo
    else:
        df = df[df > 0]
        minimo = np.nanmin(df.values)
        a = np.where(df.values == minimo)
        valor = minimo
    return [df.index[a[0][0]], df.index[a[0][1]], round(valor, 2)]
```

```python
Dedf = pd.DataFrame(De, index=df.nome, columns=df.nome)    # distância euclidiana
Depdf = pd.DataFrame(Dep, index=df.nome, columns=df.nome)   # distância euclidiana padronizada
Dmdf = pd.DataFrame(Dm, index=df.nome, columns=df.nome)     # mahalanobis
```

```python
# alunos mais semelhantes usando dedf
compara_dist(Dedf, 'min')
```

```python
# alunos mais distantes usando dedf
compara_dist(Dedf)
```

```python
# alunos mais semelhantes usando depdf
compara_dist(Depdf, 'min')
```

```python
# alunos mais distantes usando depdf
compara_dist(Depdf)
```

```python
# alunos mais semelhantes usando dmdf
compara_dist(Dmdf, 'min')
```

```python
# alunos mais distantes usando dmdf
compara_dist(Dmdf)
```
