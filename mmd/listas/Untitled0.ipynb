```python
import pandas as pd
import numpy as np
```

```python
S = np.array([7.64, 2.87, 7.04, 2.87, 37.59, 35.70, 7.04, 35.70, 44]).reshape((3, 3))
S
```

```python
np.var(S, axis=1)
```

```python
l, e = np.linalg.eig(S)
idx = np.argsort(l)[::-1]
l = l[idx]
e = e[:, idx]
print(l)
print(e)
```

```python
trace = sum(l)
trace
```

```python
np.trace(S)
```

```python
np.allclose(np.trace(S), trace)
```

```python
l[0] * l[1] * l[2]
```

```python
np.linalg.det(S)
```
