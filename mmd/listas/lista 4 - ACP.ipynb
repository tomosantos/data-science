```python
S = np.array([[5, 2], [2, 2]])
S
```

```python
l, e = np.linalg.eig(S)
idx = np.argsort(l)[::-1]
l = l[idx]
e = e[:,idx]
print(l)
```

```python
# calcular a correlação entre as duas variáveis
r12 = S[0,1] / (S[0, 0] * S[1, 1]) ** 0.5

# montar a matriz de correlações amostrais
R = np.array([[1, r12], [r12, 1]])
R
```

```python
l, e = np.linalg.eig(R)
idx = np.argsort(l)[::-1]
l = l[idx]
e = e[:,idx]
print(l)
print(e)
```
