### Indexing, Selection and Slicing

As colunas individuais de um DataFrame podem ser selecionadas com _regular indexing_. Cada coluna é representada com uma `Series`.

```python
df
```

```python
df.loc['Canada'] # olhar horizontal
```

```python
df.iloc[-1] # olhar horizontal
```

```python
df['Population'] # olhar vertical
```

Colunas múltiplas também podem ser selecionadas, assim como em `Numpy` e `Series`.

```python
df[['Population', 'GDP']]
```

Nesse caso, o resultado é outro `Dataframe`. 

O _slicing_ funciona de forma diferente, ele age em "_row level_".

```python
df[1:3]
```

"_Row Level_" funciona melhor com `loc` e `iloc`, que é mais **recomendado** do que o _slicing_ direto (`df[:]`).

`loc` seleciona as linhas que coincidem com o índice.

```python
df.loc['Italy']
```

```python
df.loc['France':'Italy']
```

Como um segundo argumento, é possível setar as colunas que você gostaria de selecionar:

```python
df.loc['France': 'Italy', 'Population']
```

```python
df.loc['France': 'Italy', ['Population', 'GDP']]
```

### Conditional Selection (_boolean arrays_)

```python
df
```

```python
df['Population'] > 70
```

```python
df.loc[df['Population'] > 70]
```

A correspondência booleana é feita em "_index level_", então, você pode filtrar por qualquer linha, contanto que contenha os índices corretos.

A seleção por coluna ainda funciona como esperado:

```python
df.loc[df['Population'] > 70, 'Population']
```

```python
df.loc[df['Population'] > 70, ['Population', 'GDP']]
```

### Dropping stuff

Se opondo à noção de "selection", temos "dropping". Ao invés de apontarmos quais valores gostaríamos de selecionar, iremos apontar para aqueles que queremos excluir:

```python
df.drop('Canada')
```

```python
df.drop(['Canada', 'Japan'])
```

```python
df.drop(columns=['Population', 'HDI'])
```

```python
df.drop(['Italy', 'Canada'], axis=0)  # definimos com 'axis=0', explicitamente, que as linhas devem ser removidas
```

```python
df.drop(['Population', 'HDI'], axis=1) # aqui, definimos, explicitamente, que as colunas devem ser removidas
```

```python
df.drop(['Population', 'HDI'], axis='columns') # no entanto, assim fica mais visível
```

```python
df.drop(['Canada', 'Germany'], axis='rows')
```

Todos esses métodos utilizando `drop` retornam um novo `DataFrame`. Caso haja a preferência para se modificar '_in place_', pode se recorrer ao métoodo `inplace`.

### Modifying DataFrames

É bem simples e intuitivo:

#### Adding a new column

```python
langs = pd.Series(
    ['French', 'German', 'Italian'],
    index=['France', 'Germany', 'Italy'],
    name='Language'
)
```

```python
langs
```

```python
df['Language'] = langs
df
```

_Repare que os valores foram preenchidos nos locais onde havia um **match** entre os índices_.

#### Replacing values per column

```python
df['Language'] = 'English'
df
```

#### Renaming Columns

```python
df.rename(
    columns={
        'HDI': 'Human Development Index',
        'Anual Popcorn Consumption': 'APC'      # repare que aqui esta coluna sequer existe, portanto, o código irá ignorar
    }, index={
        'United States': 'USA',
        'United Kingdom': 'UK',
        'Argentina': 'AR'                       # o mesmo com essa linha
    })
```

```python
df.rename(index=str.upper)
```

```python
df.rename(index=lambda x: x.lower())
```

#### Dropping Columns

```python
df.drop(columns='Language', inplace=True)
df
```

#### Adding values

```python
df.loc['China'] = pd.Series({
    'Population': 3,
    'GDP': 5
})
df
```

```python
df.loc['China'] = pd.Series({'Population': 1_400_000_000, 'Continent': 'Asia'})
df
```

```python
df.drop('China', inplace=True)
df
```

#### More radical index changes

```python
df.reset_index()
```

```python
df.set_index('Population')
```

#### Creating columns from other columns

Alterar um DataFrame quase sempre envolve combinar diferentes colunas em uma outra. Por exemplo, na nossa análise dos países, nos podemos tentar calcular o 'GDP per capita', que é, justamente o `GDP / Population`.

```python
df[['Population', 'GDP']]
```

A forma como o pandas expressa isso é justamente dividindo cada Series:

```python
df['GDP'] / df['Population']
```

O resultado da operação é uma outra Series que você pode adicionar ao `DataFrame` original:

```python
df['GDP Per Capita'] = df['GDP'] / df['Population']
```
