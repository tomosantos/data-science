## Seaborn

### Plots de distribuição

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'], kde=True)
```

```python
sns.displot(tips['total_bill'], kde=True, bins=300)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

### Plots categóricos

```python
tips
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
import numpy as np

sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='day', legend=False)
```

```python
sns.boxplot(y='day', x='total_bill', data=tips, palette='rainbow', hue='day', legend=False, orient='h')
```

```python
sns.boxplot(y='day', x='total_bill', data=tips, palette='rainbow', hue='sex', legend=False, orient='h')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='sex', split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='day')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s': 100}, aspect=1.6, height=6)
```

### Plots Matriciais

```python
flights = sns.load_dataset('flights')
flights
```

```python
df_flights_matrix = flights.pivot_table(index='year', columns='month', values='passengers')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20, 8))
sns.heatmap(df_flights_matrix, ax=ax, annot=True, fmt='.0f', linewidths=1, cmap='magma')
```

```python
sns.clustermap(df_flights_matrix)
```

### Estilização

```python
sns.set_style('white')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex', data=tips, palette='deep', hue='sex')
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex', data=tips, palette='deep', hue='sex')
```

```python
sns.set_style('dark')
sns.countplot(x='sex', data=tips, palette='deep', hue='sex')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips, palette='deep', hue='sex')
```
