## Matplotlib

### Introdução

```python
import numpy as np

x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(x, y, 'red')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Gráfico')
```

Existem dois elementos '*invisíveis*' dentro de um plot:

1. temos uma caixa imaginária que está em volta do nosso plot e ela conterá todos os gráficos que colocarmos ali dentro e
2. temos o próprio gráfico.

No matplotib, os gráficos levam o nome de **eixo** (`axe`) e a caixa imaginária leva o nome de `figure`.


```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

Com o comando `subplot` somos capazes de criar dois eixos dentro de uma mesma figure.

O método em si possui os seguintes parâmetros:
- *o número de linhas presentes na figure*
- *o número de colunas presentes na figure*
- *em qual dos eixos estamos mexendo*

### Orientação à objetos no matplotlib

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # esquerda, inferior, largura, altura

axes.plot(x, y, 'blue')
axes.set_xlabel('Label x')
axes.set_ylabel('Label y')
axes.set_title('Gráfico 1')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # esquerda, inferior, largura, altura
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  # esquerda, inferior, largura, altura

axes.plot(x, y, 'blue')
axes.set_xlabel('Label x')
axes.set_ylabel('Label y')
axes.set_title('Gráfico 1')

axes2.plot(y, x, 'red')
axes2.set_xlabel('Label x2')
axes2.set_ylabel('Label y2')
axes2.set_title('Gráfico 2')
```

### `suplots()`

```python
fig, ax = plt.subplots()

ax.plot(x, y, 'r--')
ax.set_title('Subplot Title')
```

Com `subplots()` lidamos com a noção de **desempacotamento de tuplas**.
- *Isso nos permite atribuirmos vários valores a várias variáveis de uma forma sucinta.*
- Com o método `subplots()`, os resultados são uma figura e um eixo.

Dessa forma, constantemente vemos o seguinte acontecendo:

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

for axis in ax:
    axis.plot(x, y, 'r')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x, y, 'red')
ax[1].plot(x, y, 'green')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

for axis in ax:
    axis.plot(x, y, 'r')

plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(30, 8))

for axis in ax:
    axis.plot(x, y, 'r')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
```

### Legendas, rótulos e títulos

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label='x**2', color='#FF50AB', alpha=0.5)
ax.plot(x, x**1.9, label='x**1.9', color='#FF50AB', alpha=1)
ax.plot(x, x**3, 'g*-', label='x**3')

ax.set_title('My plot')
ax.set_xlabel('Axis x')
ax.set_ylabel('Axis y')

ax.legend(loc=0)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2)
axes[0].plot(x, x**3)
axes[0].set_title('Intervalo de eixos padrão')

axes[1].plot(x, x**2)
axes[1].plot(x, x**3)
axes[1].axis('tight')
axes[1].set_title('Eixos apertados')

axes[2].plot(x, x**2)
axes[2].plot(x, x**3)
axes[2].set_ylim([0, 60]) # dá um zoom no eixo y
axes[2].set_xlim([2, 5])  # # dá um zoom no eixo x
axes[2].set_title('Intervalo de eixos padrão')
```
